# Task 2 :  Data Collection

Use nytimes github repository as a data source to collect the covid 19 data in real time and ingest it into colab notebook for data anlysis purpose

```
df = pd.read_csv('https://raw.githubusercontent.com/nytimes/covid-19-data/master/us-counties.csv')
```

# Task 3 : Data Wrangling and EDA (Exploratory Data Analysis)




```
California = df[(df['state']=='California')& (df['county'] == 'Los Angeles')]
dates = California.loc[(df['date'] >= '2021-12-21')]
print(dates)
deaths = dates.groupby(['date'])['deaths'].mean()
cases = dates.groupby(['date'])['cases'].mean()

print('\n Avg Daily Death Toll for Last 30 Days :\n',deaths,'\n Avg Daily Cases for Last 30 Days:\n',cases )
```

# Task4: Understand NY covid 19 data in last 30 days

```

NYC = df[(df['state']=='New York')& (df['county'] == 'New York City')]
date = NYC.loc[(df['date'] >= '2021-12-21')]
print(date)
death = date.groupby(['date'])['deaths'].mean()
case = date.groupby(['date'])['cases'].mean()

print('\n Avg Daily Death Toll for Last 30 Days in NYC :\n',death,'\n Avg Daily Cases for Last 30 Days in NYC:\n',case )
```

# Comparing Covid 19 data with different states

```
import pandas as pd
import seaborn as sns
df = pd.read_csv('https://raw.githubusercontent.com/nytimes/covid-19-data/master/us-counties.csv')
NYC = df[(df['state']=='New York')& (df['county'] == 'New York City')]

sns.lineplot(data = NYC, x = 'date' , y = 'cases', label="NYC COVID CASES")


LA = df[(df['state']=='California')& (df['county'] == 'Los Angeles')]

sns.lineplot(data = LA, x = 'date' , y = 'cases', label="LA COVID CASES")
```
